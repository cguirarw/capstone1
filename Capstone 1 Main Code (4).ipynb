```python
from sklearn import linear_model
```

```python
#Harsh Code: Basically the numpy does not realize the shape of the array when there is only one column. so we need to reshape
#it using reshape (-1, 1). Here -1 allows the numpy to infer the size of array. If you had multiple X column this should not
#happen

print(y.shape)
y1 = y.values.reshape(-1, 1)
print(y1.shape)
print(X.shape)
X1 = X.values.reshape(-1, 1)
print(X1.shape)
X3 = X2.values.reshape(-1, 1)
X5 = X4.values.reshape(-1, 1)
```

```python
#Setting up initial regressions for holistic statistics

lm = linear_model.LinearRegression()
model = lm.fit(X1,y1)
print(model)
lm1 = linear_model.LinearRegression()
model1 = lm1.fit(X3,y1)
lm2 = linear_model.LinearRegression()
model2 = lm2.fit(X5,y1)
```

```python
print(model.score(X1, y1))
print(model.coef_)
print(model.intercept_)
```

```python
#Setting up variables for prediction

y1_pred = lm.predict(X1)
y3_pred = lm1.predict(X3)
y5_pred = lm2.predict(X5)
```

```python
#Plotting VORP compared to WS

fig, ax = plt.subplots()
plt.plot(X1, y1_pred, color='blue', linewidth=3)
plt.scatter(X1, y, color='black')
ax.set_xlabel('Value Over Replacement Player')
ax.set_ylabel('Win Shares')
```

```python
#Plotting BPM compared to WS

fig, ax = plt.subplots()
plt.plot(X3, y3_pred, color='green', linewidth=3)
plt.scatter(X3, y, color='black')
ax.set_xlabel('Box Plus/Minus Score')
ax.set_ylabel('Win Shares')
```

```python
#Plotting PER compared to WS

fig, ax = plt.subplots()
plt.plot(X5, y5_pred, color='red', linewidth=3)
plt.scatter(X4, y, color='black')
ax.set_xlabel('Player Efficiency Rating')
ax.set_ylabel('Win Shares')
```

```python
#Plotting comparisons of holistic statistics

fig, ax = plt.subplots()
plt.plot(X1, y1_pred, color='blue', linewidth=3)
plt.plot(X3, y3_pred, color='green', linewidth=3)
plt.plot(X5, y5_pred, color='red', linewidth=3)
plt.gca().legend(('VORP','BPM','PER'))
ax.set_ylabel('Win Shares')
ax.set_xlabel('Holistic Statistic')
```

```python
#Creating function in order to separate the lower score cluster from the rest of the data for total rebounding percentage

def TRB_categories(TRB):
    
    if TRB < 8:
        return 1
    else:
        return 2
data['TRB_categories'] = data['TRB%'].apply(TRB_categories)

#Using previously created function to create plot that shows a cluster at the lower scores of total rebounding percentage

sns.lmplot(x='TRB%', y='WS', hue='TRB_categories', data=data)
plt.xlabel('Total Rebounding Percentage')
plt.ylabel('Win Shares')
```

```python
#Creating function in order to separate the lower score cluster from the rest of the data for assist percentage

def AST_categories(AST):
    
    if AST < 12:
        return 1
    else:
        return 2
data['AST_categories'] = data['AST%'].apply(AST_categories)

#Using previously created function to create plot that shows a cluster at the lower scores of assist percentage

sns.lmplot(x='AST%', y='WS', hue='AST_categories', data=data)
plt.xlabel('Assist Percentage')
plt.ylabel('Win Shares')
```
